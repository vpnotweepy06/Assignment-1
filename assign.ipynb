```python
import numpy as np

coffee["newprice"]=np.where(coffee["Coffee Type"]=="Espresso",3.99,5.99)
coffee
```

```python
# tp drop column

coffee.drop(columns = ["Price"], inplace = True) #inplace : returns modified copies
```

```python
coffee
```

```python
# revenue
coffee["revenue"] = coffee["Units Sold"] * coffee["newprice"]
coffee
```

```python
# rename

coffee.rename(columns = {"newprice":"price"}, inplace = True)
```

```python
coffee
```

```python
# lambda
coffee["Units Solds"] = coffee["price"].apply(lambda x : "Expensive" if x == 5.99 else "less expensive")
```

here we have added another column by using lambda where if the price is > 5.99,then its expensive else it is less expensive. Also used if we want something specific
